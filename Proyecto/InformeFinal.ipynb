## 3. Proceso de Simulación
### Metodología
1. **Preparación**:
   - Eliminación de columnas irrelevantes (ej. `male_female`).
   - Separación de variables numéricas/categóricas mediante funciones `getInt()`/`getCat()`.

2. **Simulación Numérica**:
   - **Distribución normal truncada**: Para variables continuas (`revenue`, `price`, etc.), usando media ± desviación estándar original.
   - **Uniforme discreta**: Para `year` (2015-2025).

3. **Simulación Categórica**:
   - Muestreo aleatorio ponderado por frecuencias observadas (ej. 28.7% probabilidad para `type = "Textured"`).

4. **Escala**:
   - Generación de **100,000 registros** (≈ 180x el dataset original).

## 5. Conclusiones
### Efectividad de la Simulación
 **Ventajas**:
- **Escalabilidad**: Generación de 100K registros sin pérdida de patrones estadísticos clave.
- **Conservación de Relaciones**: Correlaciones y distribuciones se replican adecuadamente.
- **Utilidad para Modelado**: Datos sintéticos útiles para pruebas de modelos predictivos.

**Limitaciones**:
- **Pérdida de Variabilidad Extrema**: Outliers reales no se replican fielmente.
- **Simplificación de Interacciones**: No se modelaron relaciones complejas entre variables (ej. `campaigns` → `awareness`).

### Recomendaciones
- Validar con pruebas de hipótesis (KS, ANOVA) en aplicaciones críticas.
- Combinar datos reales y simulados para mantener autenticidad en proyectos analíticos.
